```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

df = pd.read_csv("linear-regression-dataset.csv")
print("First few rows of the dataset:")
print(df.head())

# Convert to numpy arrays
X = np.array(df['experience'].values)
y = np.array(df['salary'].values)

print("\nShape of X array:", X.shape)
print("Shape of y array:", y.shape)
print("\nFirst few values of X:", X[:5].flatten())
print("First few values of y:", y[:5])
```

```python
plt.scatter(X, y, color="red", alpha=0.6, s=100)  # Changed to red with some transparency
plt.xlabel("Years of Experience")
plt.ylabel("Salary (in $1000s)")
plt.title("Scatter Plot: Experience vs Salary")
plt.grid(True, linestyle='--', alpha=0.3)  # Added light grid
plt.show()
```

```python
### These line are in a straight continous lines so we can consider using linear regression 
```

```python
# Find the correlation btw X and y

correlation = np.corrcoef(X, y)[0, 1]
print(f"Correlation coefficient between experience and salary: {correlation:.4f}")
```

```python
#Since the correlation is very close to 1, it indicates a strong positive linear relationship between years of experience and salary. This suggests that as years of experience increase, salary tends to increase as well, making linear regression a suitable model for this data.
```
